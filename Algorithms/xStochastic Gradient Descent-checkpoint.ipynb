```python
#First load the dataset by running the cell below.
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.model_selection import train_test_split

fuel = pd.read_csv('../input/dl-course-data/fuel.csv')

X = fuel.copy()
# Remove target
y = X.pop('FE')

preprocessor = make_column_transformer(
    (StandardScaler(),
     make_column_selector(dtype_include=np.number)),
    #(OneHotEncoder(sparse=False),
     (OneHotEncoder(sparse_output=True),
     make_column_selector(dtype_include=object)),
)

X = preprocessor.fit_transform(X)
y = np.log(y) # log transform target instead of standardizing

input_shape = [X.shape[1]]
print("Input shape: {}".format(input_shape))
```

### 3.Train Model

Once you've defined the model and compiled it with a loss and optimizer you're ready for training. Train the network for 200 epochs with a batch size of 128. The input data is X with target y.

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import layers
from scipy.sparse import csr_matrix

# Assuming X is a sparse matrix in CSR format
X_csr = csr_matrix(X)
dense_tensor = tf.convert_to_tensor(X_csr.todense(), dtype=tf.float32)

# Assuming your_input_shape is the shape of your input data
your_input_shape = X.shape[1]

input_layer = layers.Input(shape=(your_input_shape,), sparse=True)
model = tf.keras.Sequential([
    layers.InputLayer(input_shape=(your_input_shape,), sparse=True),
    layers.Dense(128, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1),
])

model.compile(optimizer='adam', loss='mae', metrics=['mae'])
history = model.fit(dense_tensor, y, batch_size=128, epochs=200)

q_2.check()
```
